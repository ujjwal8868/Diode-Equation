```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Constants
q = 1.602e-19  # Charge of an electron (C)
k = 1.381e-23  # Boltzmann's constant (J/K)
T = 300  # Temperature in Kelvin (Room temp)
V_T = k * T / q  # Thermal voltage (approx 25.85 mV at 300K)

# Diode parameters
I_s = 1e-12  # Reverse saturation current (A)
n = 1.5  # Ideality factor

# Function to calculate diode current
def diode_current(V):
    return I_s * (np.exp(V / (n * V_T)) - 1)

# Voltage range for plotting
V = np.linspace(-0.1, 0.7, 100)  # Voltage from -0.2V to 0.7V
I = diode_current(V)  # Calculate current

# Plot the diode I-V characteristic
plt.figure(figsize=(8, 5))
plt.plot(V, I, label="Diode Current", color='r')
plt.axhline(0, color='black', linewidth=0.5)  # x-axis
plt.axvline(0, color='black', linewidth=0.5)  # y-axis
plt.xlabel("Voltage")
plt.ylabel("Current")
plt.title("Shockley Diode I-V Characteristic")
plt.legend()
plt.grid()
plt.yscale("log")  # Log scale for better visualization of small currents
plt.show()
```
